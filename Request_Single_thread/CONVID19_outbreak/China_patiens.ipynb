```python
import requests
import json

import sys
sys.path.append('D:\Anaconda3\Lib\site-packages')
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Geo, Map

from pyecharts.globals import ChartType, SymbolType
```

```python
url ="https://view.inews.qq.com/g2/getOnsInfo?name=disease_h5"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36',}
response = requests.get(url, headers = headers)
data = json.loads(response.json()['data']) # 将结果转化为字典
```

```python
'''json.dumps将结果转化为字符串'''
# print(json.dumps(r)) # 一行显示
# print(json.dumps(r,indent=1)) # 按照json格式显示
```

```python
time = data['lastUpdateTime']
china = data['areaTree'][0]['children'] # 获取中国各省份的数据
prince_num = len(china) # 获取中国的省份个数

print(china[0]['total'])
princeDara = {}
for i in range(prince_num):
    princeDara[china[i]['name']] = [china[i]['total']['nowConfirm'],# 现有确诊人数
                                    china[i]['total']['confirm'],# 总的确诊人数
                                    china[i]['total']['dead'] # 死亡人数
                                    ]
```

```python
total_confirm = [[k,v[1]] for k,v in princeDara.items()]
now_confirm = [[k,v[0]] for k,v in princeDara.items()]
print(total_confirm)
```

```python
# 分段图例
def map_visualmap_piece() -> Map:
    c = (
        Map()
        .add("确诊总人数", total_confirm, "china")
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="中国疫情地图"),
            visualmap_opts=opts.VisualMapOpts(pieces= [
                                 {"min": 5000},
                                 {"min": 1000, "max": 4999},
                                 {"min": 500, "max": 999},
                                 {"min": 200, "max": 499},
                                 {"min": 10, "max": 199},
                                 {"max": 9}], split_number=8, is_piecewise=True),
        )
    )
    return c
# if __name__ == '__main__':
map_piece = map_visualmap_piece()
# map_piece.render(path="china_confirm.html") # 保存
map_piece.render_notebook()
```
