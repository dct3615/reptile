# 获取各国的数据

```python
import requests
import json
import jsonpath
```

```python
# 爬取的网站
url ="https://api.inews.qq.com/newsqa/v1/automation/foreign/country/ranklist"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36',}
response = requests.get(url, headers = headers)
text = json.loads(response.text) # 数据转化为字典

# 获取数据
name = jsonpath.jsonpath(text, '$..name')
confirm = jsonpath.jsonpath(text, '$..confirm')
print(len(name),len(confirm))
```

# 数据可视化

```python
import sys
sys.path.append('D:\Anaconda3\Lib\site-packages')
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Geo, Map

from pyecharts.globals import ChartType, SymbolType
```

```python
import pickle
# 获取中文英文映射
f = open('nameMap', 'rb')
nameMap = pickle.load(f)
```

```python
data = zip(name, confirm) # 数据打包

map = Map(opts.InitOpts(width = '1000px',height = '600px')).add(series_name = '世界疫情分布',
                data_pair = data, # 输入数据
                maptype = "world", # 世界地图
                name_map = nameMap,
                is_map_symbol_show = False # 关闭标记点
               ) 
map.set_series_opts(label_opts=opts.LabelOpts(is_show=False)) # 不显示国家的名字
map.set_global_opts(
            title_opts=opts.TitleOpts(title="世界疫情地图"),
            visualmap_opts=opts.VisualMapOpts(pieces= [
                                 {"min": 200000},
                                 {"min": 100000, "max": 199999},
                                 {"min": 50000, "max": 99999},
                                 {"min": 10000, "max": 49999},
                                 {"min": 5000, "max": 9999},
                                 {"min": 1000, "max": 4999},
                                 {"max": 999}], split_number=8, is_piecewise=True),
        )
# map.render('世界疫情分布图.html')
map.render_notebook()
```
